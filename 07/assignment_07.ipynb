## import library

```python
import numpy as np
import matplotlib.image as img
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from mpl_toolkits.mplot3d import Axes3D
```

## load point data for training and testing

```python
fname_data  = 'assignment_07_data.csv'

data        = np.genfromtxt(fname_data, delimiter=',')
num_data    = data.shape[0]

xx          = np.zeros(num_data)
yy          = np.zeros(num_data)
zz          = np.zeros(num_data)

for i in range(num_data):

    xx[i] = data[i,0]
    yy[i] = data[i,1]
    zz[i] = data[i,2]
    
print(data.shape)
print(xx.shape)
print(xx.shape[0])
```

## plot the data in the three dimensional space

```python
fig = plt.figure(figsize=(12, 8))

ax1 = plt.subplot(111, projection='3d')
plt.title('data')

ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.set_zlabel('$z$')

ax1.scatter(xx, yy, zz, marker='o', color='blue', alpha=0.2)

plt.tight_layout()
plt.show()
```

## compute the loss function

```python
def compute_residual(theta, x, y, z):
    # ===== FILL UP THE CODE =====
    residual = np.zeros(num_data)
    
    for i in range(num_data):
        residual[i] = (theta[0] + (theta[1] * x[i]) + (theta[2] * y[i])) - z[i]

    # ============================
    return residual, num_data
```

```python
def compute_loss(theta, x, y, z):
    # ===== FILL UP THE CODE =====
    
    result = 0
    
    (residual, num_data) = compute_residual(theta, x, y, z)
    
    for i in range(num_data):
        residual_i = residual[i]
        power_residual_i = np.power(residual_i, 2)
        result = result + power_residual_i
        
    loss = result / 2 / num_data

    # ============================
    return loss
```

## compute the gradient for each model parameter (DO NOT COMPUTE THE GRADIENT FOR EACH MODEL PARAMETER, BUT DO COMPUTE THE GRADIENT OF THE MODEL PARAMTER VECTOR)

```python
def compute_gradient(theta, x, y, z):
    # ===== FILL UP THE CODE =====
    sum_residual = np.zeros(3)
    (residual, num_data) = compute_residual(theta, x, y, z)
    
    for i in range(num_data):
        temp = np.array((1, x[i], y[i]))
        sum_residual = sum_residual + np.multiply(residual[i], temp)
        
    grad = sum_residual / num_data

    # ============================
    return grad
```

## gradient descent for each model parameter

```python
num_iteration   = 1000
learning_rate   = 0.01

theta           = np.array((0, 0, 0))
theta_iteration = np.zeros((num_iteration, theta.size))
loss_iteration  = np.zeros(num_iteration)

for i in range(num_iteration):
    # ===== FILL UP THE CODE =====
    theta = theta - (learning_rate * compute_gradient(theta, xx, yy, zz))
    loss = compute_loss(theta, xx, yy, zz)
    
    theta_iteration[i] = theta
    loss_iteration[i] = loss
    
    # ============================
    print("iteration = %4d, loss = %5.5f" % (i, loss))
```

## plot the results

```python
def plot_loss_curve(loss_iteration):

    plt.figure(figsize=(8,6))
    plt.title('loss')
    # ===== FILL UP THE CODE =====

    index = np.arange(num_iteration)
    plt.plot(index, loss_iteration, '-', color = 'red')

    # ============================
    plt.tight_layout()
    plt.show()
```

```python
def plot_data(xx, yy, zz):
    
    fig = plt.figure(figsize=(12, 8))
    ax  = plt.subplot(111, projection='3d')
    plt.title('data')
    # ===== FILL UP THE CODE =====

    ax.scatter(xx,yy,zz, 'o', color = 'blue', alpha = 0.2)

    # ============================    
    plt.tight_layout()
    plt.show()
```

```python

def plot_model_parameter(theta_iteration):

    plt.figure(figsize=(8,6))
    plt.title('model parameter')
    # ===== FILL UP THE CODE =====

    index = np.arange(num_iteration)
    plt.plot(index, theta_iteration[:, 0], '-', color = 'red')
    plt.plot(index, theta_iteration[:, 1], '-', color = 'green')
    plt.plot(index, theta_iteration[:, 2], '-', color = 'blue')

    # ============================
    plt.tight_layout()
    plt.show()
```

```python
plot_loss_curve(loss_iteration)
```

```python
plot_model_parameter(theta_iteration)
```

```python
# ===== FILL UP THE CODE =====
X0 = np.arange(-10, 10, 0.1)
Y0 = np.arange(-10, 10, 0.1)

XX, YY = np.meshgrid(X0, Y0)
ZZ = theta[0] + (theta[1] * XX) + (theta[2] * YY)
# ============================

def plot_surface(XX, YY, ZZ, xx, yy, zz):

    fig = plt.figure(figsize=(8,8))
    ax  = fig.add_subplot(111, projection='3d')
    plt.title('loss surface')

    ax  = plt.axes(projection='3d')
    # ===== FILL UP THE CODE =====

    ax.plot_surface(XX, YY, ZZ, color = 'green')
    ax.scatter(xx, yy, zz, marker = 'o', color = 'blue', alpha = 0.2)

    # ============================
    plt.tight_layout()
    plt.show()
```

```python
plot_surface(XX, YY, ZZ, xx, yy, zz)
```
