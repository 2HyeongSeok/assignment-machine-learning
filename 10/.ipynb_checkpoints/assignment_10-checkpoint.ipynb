# Binary classification based on Logistic Regression with a quadratic regularization

## import library

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

## load data

```python
fname_data_train    = 'assignment_10_data_train.csv'
fname_data_test     = 'assignment_10_data_test.csv'

data_train  = np.genfromtxt(fname_data_train, delimiter=',')
data_test   = np.genfromtxt(fname_data_test, delimiter=',')

num_train = data_train.shape[0]
num_test = data_test.shape[0]

train_x = np.zeros(num_train)
train_y = np.zeros(num_train)
label_train = np.zeros(num_train)

test_x = np.zeros(num_test)
test_y = np.zeros(num_test)
label_test = np.zeros(num_test)

for i in range(num_train):
    train_x[i] = data_train[i,0]
    train_y[i] = data_train[i,1]
    label_train[i] = data_train[i,2]
    
for i in range(num_test):
    test_x[i] = data_test[i,0]
    test_y[i] = data_test[i,1]
    label_test[i] = data_test[i,2]
```

## plot the data

```python
train_x0 = []
train_x1 = []
train_y0 = []
train_y1 = []

test_x0 = []
test_x1 = []
test_y0 = []
test_y1 = []

for i in range(num_train):
    if label_train[i] == 0:
        train_x0.append(train_x[i])
        train_y0.append(train_y[i])
    else:
        train_x1.append(train_x[i])
        train_y1.append(train_y[i])
        
for i in range(num_test):
    if label_test[i] == 0:
        test_x0.append(test_x[i])
        test_y0.append(test_y[i])
    else:
        test_x1.append(test_x[i])
        test_y1.append(test_y[i])
        
plt.figure(figsize=(16,8))
            
plt.subplot(121)
plt.title('training data')
plt.scatter(train_x0, train_y0, marker = '.', s = 30, color = 'blue', label = 'class 0')
plt.scatter(train_x1, train_y1, marker = '.', s = 30, color = 'red', label = 'class 1')
plt.legend()
plt.tight_layout()

plt.subplot(122)
plt.title('testing data')
plt.scatter(test_x0, test_y0, marker = '.', s = 30, color = 'blue', label = 'class 0')
plt.scatter(test_x1, test_y1, marker = '.', s = 30, color = 'red', label = 'class 1')
plt.legend()
plt.tight_layout()
```

## define feature function

```python
def feature_function(x, y):
    xx = x**2
    xxx = x**3
    
    feature = np.array((1,x,y,xx,xxx), dtype = object)
    return feature
```

## define regression function based on the feature function

```python
def regression_function(theta, feature):
    value = np.matmul(theta, feature)
    return value
```

## define regularization function on the model parameters

```python
def regularization_function(theta):
    square = np.square(theta)
    value = np.sum(square)
    return value
```

## define sigmoid function

```python
def logistic_function(x):
    exp_temp = 1 + np.exp(-x)
    z = 1 / exp_temp
    return z
```

## define loss function where $\alpha$ is a weight for the quadratic regularization term (Note that you need to add a small number (np.finfo(float).eps) inside logarithm function in order to avoid $\log( 0 )$)

```python
def compute_loss_feature(theta, feature, label, alpha):
    loss = 0
    for i in range(label.shape[0]):
        hi = logistic_function(regression_function(theta, feature[:, i]))
        a = (-1) * label[i] * np.log(hi + np.finfo(float).eps)
        b = (-1) * (1 - label[i]) * np.log(1 - hi + np.finfo(float).eps)
        quad = alpha * regularization_function(theta) / 2
        
        loss_temp = a + b + quad
        loss = loss + loss_temp
        
    loss = loss / label.shape[0]
    return loss
```

## define gradient vector for the model parameters with the quadratic regularization term whose weight is $\alpha$

```python
def compute_gradient_feature(theta, feature, label, alpha):
    gradient = 0
    for i in range(label.shape[0]):
        a_diff = label[i] * (logistic_function(regression_function(theta, feature[:, i])) - 1)
        b_diff = (1 - label[i]) * logistic_function(regression_function(theta, feature[:, i]))
        quad_diff = alpha * theta
        
        gradient_temp = np.multiply(a_diff, feature[:, i]) + np.multiply(b_diff, feature[:, i]) + np.multiply(quad_diff, feature[:, i])
        gradient = gradient + gradient_temp
        
    gradient = gradient / label.shape[0]
    return gradient
```

## compute the accuracy

```python
def compute_accuracy(theta, feature, label):
    count = 0
    newlabel = np.zeros(label.shape[0])
    
    for i in range(label.shape[0]):
        value = regression_function(theta, feature[:, i])
        
        if value > 0:
            newlabel[i] = 1
        else:
            newlabel[i] = 0
    for i in range(label.shape[0]):
        if label[i] == newlabel[i]:
            count += 1
            
    accuracy = count / label.shape[0]
    return accuracy
```

## gradient descent for the model parameters $\theta$

```python
num_iteration   = 10000
learning_rate   = 0.7
alpha           = 0.00000001

theta = np.array((0,0,0,0,0))
```

```python
theta_iteration             = np.zeros((num_iteration, theta.size))
loss_iteration_train        = np.zeros(num_iteration)
loss_iteration_test         = np.zeros(num_iteration)
accuracy_iteration_train    = np.zeros(num_iteration)
accuracy_iteration_test     = np.zeros(num_iteration)

feature_train_init = np.array([1, train_x[0], train_y[0], (train_x[0])**2, (train_x[0])**3])
for i in range(1, num_train):
    tuple = feature_function(train_x[i], train_y[i])
    feature_train_init = np.vstack([feature_train_init, tuple])
feature_train = np.transpose(feature_train_init)

feature_test_init = np.array([1, test_x[0], test_y[0], (test_x[0])**2, (test_x[0])**3])
for i in range(1, num_train):
    tuple = feature_function(train_x[i], train_y[i])
    feature_test_init = np.vstack([feature_test_init, tuple])
feature_test = np.transpose(feature_test_init)
```

```python
for i in range(num_iteration):
    theta = theta - (learning_rate * compute_gradient_feature(theta, feature_train, label_train, alpha))
    loss_train = compute_loss_feature(theta, feature_train, label_train, alpha)
    loss_test = compute_loss_feature(theta, feature_test, label_test, alpha)
    accuracy_train = compute_accuracy(theta, feature_train, label_train)
    accuracy_test = compute_accuracy(theta, feature_test, label_test)
    
    loss_iteration_train[i] = loss_train
    loss_iteration_test[i] = loss_test
    accuracy_iteration_train[i] = accuracy_train
    accuracy_iteration_test[i] = accuracy_test
    theta_iteration[i] = theta
    print("[%04d] loss(train) = %5.5f, loss(test) = %5.5f, accuracy(train) = %5.5f, accuracy(test) = %5.5f" % (i, loss_train, loss_test, accuracy_train, accuracy_test))

    
theta_optimal = theta
```

## compute accuracy of the classifiers

```python
accuracy_train  = compute_accuracy(theta_optimal, feature_train, label_train)
accuracy_test   = compute_accuracy(theta_optimal, feature_test, label_test)
```

## plot the results

```python
def plot_loss_curve(loss_iteration_train, loss_iteration_test):

    plt.figure(figsize=(8,6))   # USE THIS VALUE for the size of the figure
    plt.title('loss')
    
    index = np.arange(num_iteration)
    plt.plot(index, loss_iteration_train, '-', color = 'red')
    plt.plot(index, loss_iteration_test, '-', color = 'blue')
    
    plt.xlabel('iteration')
    plt.ylabel('loss')
    plt.tight_layout()
    plt.show()
```

```python
def plot_accuracy_curve(accuracy_iteration_train, accuracy_iteration_test):

    plt.figure(figsize=(8,6))   # USE THIS VALUE for the size of the figure
    plt.title('accuracy')
    
    index = np.arange(num_iteration)
    plt.plot(index, accuracy_iteration_train, '-', color = 'red')
    plt.plot(index, accuracy_iteration_test, '-', color = 'blue')
    
    plt.xlabel('iteration')
    plt.ylabel('loss')
    plt.tight_layout()
    plt.show()
```

```python
def plot_data(data_train, data_test):
    num_train = data_train.shape[0]
    num_test = data_test.shape[0]
    
    train_x = np.zeros(num_train)
    train_y = np.zeros(num_train)
    label_train = np.zeros(num_train)

    test_x = np.zeros(num_test)
    test_y = np.zeros(num_test)
    label_test = np.zeros(num_test)

    for i in range(num_train):
        train_x[i] = data_train[i,0]
        train_y[i] = data_train[i,1]
        label_train[i] = data_train[i,2]

    for i in range(num_test):
        test_x[i] = data_test[i,0]
        test_y[i] = data_test[i,1]
        label_test[i] = data_test[i,2]
    
    train_x0 = []
    train_x1 = []
    train_y0 = []
    train_y1 = []

    test_x0 = []
    test_x1 = []
    test_y0 = []
    test_y1 = []

    for i in range(data_train.shape[0]):
        if label_train[i] == 0:
            train_x0.append(train_x[i])
            train_y0.append(train_y[i])
        else:
            train_x1.append(train_x[i])
            train_y1.append(train_y[i])

    for i in range(num_test):
        if label_test[i] == 0:
            test_x0.append(test_x[i])
            test_y0.append(test_y[i])
        else:
            test_x1.append(test_x[i])
            test_y1.append(test_y[i])
    
    f = plt.figure(figsize=(16,8))   

    plt.subplot(121)
    plt.title('training data')
    plt.scatter(train_x0, train_y0, marker = '.', s = 30, color = 'blue', label = 'class 0')
    plt.scatter(train_x1, train_y1, marker = '.', s = 30, color = 'red', label = 'class 1')
    plt.legend()
    plt.tight_layout()

    plt.subplot(122)
    plt.title('testing data')
    plt.scatter(test_x0, test_y0, marker = '.', s = 30, color = 'blue', label = 'class 0')
    plt.scatter(test_x1, test_y1, marker = '.', s = 30, color = 'red', label = 'class 1')
    plt.legend()
    plt.tight_layout()
```

```python
def plot_model_parameter(theta_iteration):
    plt.figure(figsize=(8,6))   # USE THIS VALUE for the size of the figure
    plt.title('model parameter')
    index = np.arange(num_iteration)
    
    plt.plot(index, theta_iteration[:, 0], '-', color = 'red', label = 'theta 0')
    plt.plot(index, theta_iteration[:, 1], '-', color = 'green', label = 'theta 1')
    plt.plot(index, theta_iteration[:, 2], '-', color = 'blue', label = 'theta 2')
    plt.plot(index, theta_iteration[:, 3], '-', color = 'orange', label ='theta 3')
    plt.plot(index, theta_iteration[:, 4], '-', color = 'gray', label = 'theta 4')
             
    plt.xlabel('iteration')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
def plot_classifier(data, theta):
    x = np.zeros(data.shape[0])
    y = np.zeros(data.shape[0])
    l = np.zeros(data.shape[0])
    
    for i in range(data.shape[0]):
        x[i] = data[i,0]
        y[i] = data[i,1]
        l[i] = data[i,2]
    
    xx1 = np.arange(x.min() - 1, x.max() + 1, 0.01)
    yy1 = np.arange(y.min() - 1, y.max() + 1, 0.01)
    xx10, yy10 = np.meshgrid(xx1, yy1)
    
    feature10 = feature_function(xx10, yy10)
    Z1 = regression_function(theta,feature10)

    plt.figure(figsize=(8,8)) # USE THIS VALUE for the size of the figure
    
    point_x0 = []
    point_y0 = []
    point_x1 = []
    point_y1 = []
    for i in range(l.shape[0]):
        if l[i] == 0:
            point_x0.append(x[i])
            point_y0.append(y[i])
        else:
            point_x1.append(x[i])
            point_y1.append(y[i])
        
    contour = plt.contourf(xx10, yy10, Z1, levels = 70, cmap = plt.cm.get_cmap('RdBu_r'))
    ax_c = plt.colorbar(contour)
    plt.scatter(point_x0, point_y0, s = 30, color = 'blue', label = 'class 0', edgecolor="white")
    plt.scatter(point_x1, point_y1, s = 30, color = 'red', label = 'class 1', edgecolor="white")
    plt.contour(xx10, yy10, Z1, levels= [.8], cmap="Greys", vmin = 0, vmax = .6)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
